```python
from tqdm import tqdm
import textacy
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
import string
import re
import spacy
```

```python
stopWords = set(stopwords.words('english'))
nlp = spacy.load('en')
stopWords = set(stopwords.words('english'))
```

```python
train = pd.read_csv("train.csv")
```

```python

#replace with space
def preprocess(x):
    x = re.sub("won't","will not",x)
    x = re.sub("cant't","can not",x)
    x = re.sub("cannot","can not",x)
    x = re.sub("n't"," not",x)
    x = re.sub("what's","what is",x)
    x = re.sub("'ve"," have",x)
    x = re.sub("i'm","i am",x)
    x = re.sub("'re"," are",x)
    x = re.sub("he's","he is",x)
    x = re.sub("she's","she is",x)
    x = re.sub("e.g","example",x)
    x = re.sub("'ll"," will",x)
    x = re.sub('['+string.punctuation+']', ' ', str(x))
    x = textacy.preprocess.normalize_whitespace(x)
    return x
    

 
```

```python
train.question1 = train.question1.apply(lambda x: textacy.preprocess.preprocess_text(str(x),
                                                                                     fix_unicode=True,
                                                                                     lowercase=True,
                                                                                     no_currency_symbols=True,
                                                                                     no_contractions=True,
                                                                                     no_punct = False,
                                                                                     no_accents = True))
```

```python
train.question2 = train.question2.apply(lambda x: textacy.preprocess.preprocess_text(str(x),
                                                                                     fix_unicode=True,
                                                                                     lowercase=True,
                                                                                     no_currency_symbols=True,
                                                                                     no_contractions=True,
                                                                                     no_punct = False,
                                                                                     no_accents = True))
```

```python
train.question1 = train.question1.apply(lambda x:preprocess(x))
train.question2 = train.question2.apply(lambda x:preprocess(x))
```

```python
q1 = []
for doc in tqdm(nlp.pipe(train['question1'], n_threads=8, batch_size=10000)):
    word_list = ([c.lemma_ for c in doc if c.lemma_ not in stopWords])
    q1.append(' '.join(i for i in word_list))
```

```python
q2 = []
for doc in tqdm(nlp.pipe(train['question2'], n_threads=8, batch_size=10000)):
    word_list = ([c.lemma_ for c in doc if c.lemma_ not in stopWords])
    q2.append(' '.join(i for i in word_list))
```

```python
q1 = pd.DataFrame(q1)
q2 = pd.DataFrame(q2)
```

```python
train['q1'] = q1
train['q2'] = q2
```

```python
train.to_csv('processedTrain.csv', index = False)
```
